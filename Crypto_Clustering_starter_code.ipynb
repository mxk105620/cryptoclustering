### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    
    # 2. Fit the model to the data using `df_market_data_scaled`
    kmeans_model.fit(df_market_data_scaled)
    
    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans_model.inertia_)

print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k_values': k_values, 'inertia_values': inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
print(elbow_df.head())
```

```python
# Plot a line chart with all the inertia values computed with 

# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(x='k_values', y='inertia_values', title='Elbow Curve', xlabel='Number of Clusters (k)', ylabel='Inertia', line_color='purple', line_width=2)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves


# Function to compute inertia for a given k
def compute_inertia(data, k):
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    kmeans_model.fit(data)
    return kmeans_model.inertia_

# Function to create Elbow curve plot
def plot_elbow_curve(data, title, xlabel, ylabel, color, line_width):
    k_values = list(range(1, 12))
    inertia_values = [compute_inertia(data, k) for k in k_values]

    elbow_df = pd.DataFrame({'k_values': k_values, 'inertia_values': inertia_values})
    elbow_plot = elbow_df.hvplot.line(x='k_values', y='inertia_values', title=title, xlabel=xlabel, ylabel=ylabel, line_color=color, line_width=line_width)

    return elbow_plot

# Create Elbow curve plots for both datasets
elbow_plot_original = plot_elbow_curve(df_market_data_scaled, 'Elbow Curve (Original Data)', 'Number of Clusters (k)', 'Inertia', 'blue', 2)
elbow_plot_pca = plot_elbow_curve(pca_data_df, 'Elbow Curve (PCA Data)', 'Number of Clusters (k)', 'Inertia', 'orange', 2)

# Display the composite plot
composite_plot = elbow_plot_original * elbow_plot_pca
composite_plot.opts(legend_position='top_right', width=800, height=500)
```
